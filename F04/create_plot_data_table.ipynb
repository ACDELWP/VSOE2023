```python
from pandas import read_csv
```

```python
data = read_csv('ForestFrag_2022.csv')
```

```python
data.iloc[0]
```

```python
data.drop(
    index=data.loc[data.GENER_DESC == 'MARINE'].index.tolist(), 
    inplace=True
)
```

```python
data.drop(
    index=data.loc[data.GENER_DESC.isnull() == True].index.tolist(), 
    inplace=True
)
```


Bioregion | Gridcode (1-5) | MMTGEN (plm 25) | Area Ha 

Grid code 1 - Patch | 2 - Transitional | 3 - Perforated | 4 - Edge | 5 - Interior | 0=nonforest 



```python
table_data = data[['bioregion', 'gridcode', 'MMTGEN', 'AREA_HA']]
```

```python
table_data.head()
```

```python
table_data.bioregion.value_counts()
```

```python
table_data.drop(
    index=table_data.loc[table_data.gridcode == 0].index.tolist(), 
    inplace=True
)
```

```python
table_data.drop(
    index=table_data.loc[table_data.bioregion.isnull() == True].index.tolist(), 
    inplace=True
)
```

```python
#table_data.drop(
#    index=table_data.loc[table_data.MMTGEN.isnull() == True].index.tolist(), 
#    inplace=True
#)
# MMTGEN.isnull() should be private
```

```python
table_data.loc[table_data.loc[table_data.MMTGEN.isnull() == True].index.tolist(), 'MMTGEN'] = 'PRIVATE'
```

```python
table_plot_data_name = 'table_plot_data.csv'
```

```python
with open(table_plot_data_name, 'w') as w:
    header = 'bioregion,gridcode,MMTGEN,local_area,local_area_pct,sum_area\n'
    w.writelines(header)
```

```python
for bioregion in table_data.bioregion.unique():
    subset = table_data.loc[table_data.bioregion == bioregion]
    sum_area = subset.AREA_HA.sum()
    for gridcode in table_data.gridcode.unique():
        for MMTGEN in table_data.MMTGEN.unique():
            local_area = subset.loc[(subset.gridcode == gridcode) & 
                                        (subset.MMTGEN == MMTGEN)
                                       ].AREA_HA.sum()
            # 'bioregion,gridcode,gridcode_name,MMGTGEN,local_area,local_area_pct,sum_area\n'
            output_line = f'{bioregion},{gridcode},{MMTGEN},{local_area},{local_area/sum_area},{sum_area}\n'
            with open(table_plot_data_name, 'a') as a:
                a.writelines(output_line)
                print(output_line)
    
```
